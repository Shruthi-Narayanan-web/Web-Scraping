```python
#import libraries
from bs4 import BeautifulSoup
#used to parse through HTML format and lets you navigate the document's structure easily and extract data
import requests
#optional to send email
import csv
```

```python
headers=({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36 OPR/113.0.0.0', 'Accept-Language': 'en-US, en;q=0.5'})
```

```python
url = 'https://www.orangeskies.in/collections/mugs-and-cups?srsltid=AfmBOoq4KHDUp1_aUaAX2cd-OhUGUSJ3UROLXzAunMAbeomc1DXtUavY'
```

```python
res=requests.get(url,headers=headers)
print(res)
```

```python
s= BeautifulSoup(res.content,'html.parser')

s1= BeautifulSoup(s.prettify(), 'html.parser')
print(s1)
```

```python
links = list(s.find_all('a', attrs={'class':'full-unstyled-link'}))
L=[]
for j in range(len(links)):
   product_list='https://www.orangeskies.in'+ links[j].get('href')
   new_webpage=requests.get(product_list,headers=headers)
   n_s=BeautifulSoup(new_webpage.content,'html.parser')
   t=n_s.find("a",attrs={'class':'product__title'}).text.strip()
   p=n_s.find("span",attrs={'class':'price-item price-item--sale price-item--last'}).text.strip()
   m=n_s.find("div",attrs={'class':'made-in-india'}).text.strip()
   d=n_s.find("span",attrs={'class':'badge price__badge-sale color-accent-2'}).text.strip()
   L.append([t,p,m,d])
```

```python
print(L)
```

```python
print('Title', 'Price','Country of manufacture', 'Discount', sep='\t')
for i in L:
  for u in i:
    print(u, sep='    ', end=" ")
  print()
```
